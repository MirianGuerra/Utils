```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dic = {'Convencional':[47.3], 'Terapia Alvo': [80.8]}
df = pd.DataFrame(dic)
df.plot(kind='bar', legend=True)
plt.title('Gr√°fico de Barras - Convencional vs Terapia Alvo')
plt.ylabel('Valores')
plt.xlabel('Categorias')
plt.show()
```

```python
dic = {'Convencional':[47.3], 'Terapia Alvo': [80.8]}
df = pd.DataFrame(dic)
sns.barplot(data=df, palette="viridis")
# Adicione os valores acima das barras
for index, row in df.iterrows():
    plt.text(index, row['Convencional']+1, str(row['Convencional']), color='black', ha='center')
    plt.text(index+1, row['Terapia Alvo']+1, str(row['Terapia Alvo']), color='black', ha='center')
    
plt.title('Sobrevida para Leucemia Miel√≥ide Cr√¥nica - Global')
# Remova o quadro (frame) ao redor do gr√°fico
sns.despine()

plt.ylim([0,100])
plt.show()
```
